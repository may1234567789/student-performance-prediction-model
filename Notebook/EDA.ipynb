```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

plt.style.use("ggplot")
```

```python
df = pd.read_excel(r"C:\Users\mayank\Desktop\Data Science\Dataset\student_performance_dataset_enhanced.xlsx")
df.head()
df.shape
df.columns
df.info()
df.describe()
df.isnull().sum()
```

```python
df["Final_Grade"].value_counts()

sns.countplot(data=df, x="Final_Grade")
plt.title("Grade Distribution")
plt.show()

df.hist(figsize=(15,10))
plt.show()

numeric_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(10,8))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()

numeric_df.boxplot(figsize=(15,8))
plt.xticks(rotation=45)
plt.show()

sns.pairplot(
    df,
    hue="Final_Grade",
    vars=[
        "Attendance",
        "Study_Hours",
        "Assignments_Completed",
        "Previous_Exam_Marks"
    ]
)
```
